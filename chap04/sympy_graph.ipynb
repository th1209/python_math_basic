#### SymPyによるグラフ描画サンプル
* 以下に、簡単なグラフの描画例を示す。
* SymPyを使うと、プログラム側でプロットすべき点の座標を作らなくても良いので便利。

```python
from sympy.plotting import plot
from sympy import Symbol
```

```python
x = Symbol('x')
```

```python
# 一次関数を描画する
plot((2*x + 3), (x, -5, 5), title='Draw primary function')
```

```python
# 二次関数を描画する
# (rangeはxlim, ylimでも設定可能。第二引数に指定するより、こちらの方が便利かもしれない。)
plot(x**2 -4 * x + 5,  title='Draw quadratic function', xlim=(-5, 7), ylim=(0, 10))
```

```python
# ２つ以上の関数を描画する
p = plot(2*x + 3, 3*x + 1,  legend=True, show= False)
p[0].line_color='b'
p[1].line_color='r'
p.show()
```
