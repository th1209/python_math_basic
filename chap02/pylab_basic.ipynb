#### pylabの基本

```python
from pylab import plot
from pylab import show
from pylab import title
from pylab import ylabel
from pylab import xlabel
from pylab import legend
from pylab import axis
```

#### 基本的な点のプロット

```python
x_numbers = [1, 2, 3]
y_numbers = [2, 4, 6]


plot(x_numbers, y_numbers, marker='o')
show()
```

#### NYCの13年分の気温を描画する

```python
nyc_temp = [53.9, 56.3, 56.4, 53.4, 54.5, 55.8, 56.8, 55.0, 55.3, 54.0, 56.7, 56.4, 57.3]
years = range(2000, 2013)
plot(years, nyc_temp, marker='x')
show()
```

#### NYCの各年の気温を描画する(複数グラフの表示、グラフの表示の調整)

```python
nyc_temp_2000 = [31.3, 37.3, 47.2, 51.0, 63.5, 71.3, 72.3, 72.7, 66.0, 57.0, 45.3, 31.1]
nyc_temp_2006 = [40.9, 35.7, 43.1, 55.7, 63.1, 71.0, 77.9, 75.8, 66.6, 56.2, 51.9, 43.6]
nyc_temp_2012 = [37.3, 40.9, 50.9, 55.8, 65.1, 71.0, 77.8, 76.8, 68.8, 58.0, 43.9, 41.5]
months = range(1, 13)

# plotに、x軸・y軸のリストの対を順に渡すと、
# 複数のグラフをまとめて描画できる。
plot(months, nyc_temp_2000, months, nyc_temp_2006, months, nyc_temp_2012)

# 勿論、別個にplotを呼び出してもOK。
# plot(months, nyc_temp_2000)
# plot(months, nyc_temp_2006)
# plot(months, nyc_temp_2012)

# 以下のように、グラフの表示を調整できる。
title('Average monthly temperature in NYC')
xlabel('Month')
ylabel('Temperature')
legend([2000, 2006, 2012])
axis(ymin=30, ymax=100)

show()
```
